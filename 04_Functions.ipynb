## Example1: Implemetion of the below relation:(without function)


```python
c,k=input().split()
c=int(c)
k=int(k)
j=1
a=1
while j<=c:
    a*=j
    j+=1
i=1
b=1
while i<=k:
    b*=i
    i+=1
i=1
d=1
while i<=c-k:
    d*=i
    i+=1
print(a/(b*d))
```

## Example2:Solve the pervious question by using function

```python
def fact(c):
    a=1
    j=1
    while j<=c:
        a*=j
        j+=1
    return a
c,k=input().split()
c=int(c)
k=int(k)
a=fact(c)
b=fact(k)
d=fact(c-k)
print(a/(b*d))
```

## Example3: Computing average for students


```python
def getAvg(n):
    i=1
    Sum=0
    while i<=n:
        g=float(input('>'))
        Sum+=g
        i+=1
    return Sum/n
Maxavg=0
while True:
    n=int(input('Number of courses:'))
    if n<=0:
        break
    avg=getAvg(n)
    if avg>Maxavg:
        Maxavg=avg
print('Average of the TOP student:',Maxavg)
```

```python
def getAvg(n):
    i=1
    Sum=0
    SumImpact=0
    while i<=n:
        g=float(input('grade>'))
        impact=float(input('impact>'))
        Sum+=g*impact
        SumImpact+=impact
        i+=1
    print('The Average:',Sum/SumImpact)
    return Sum/SumImpact
Maxavg=0
while True:
    n=int(input('Number of courses:'))
    if n<=0:
        break
    avg=getAvg(n)
    if avg>Maxavg:
        Maxavg=avg
print('Average of the TOP student:',Maxavg)
```

## Example4:Score of Students


```python
def getBSc(avg,nC):
    return avg*0.8+nC*0.2
def getMSc(avg, nP):
    return avg*0.6+nP*0.4
Bscore=0
nb=0
MaxB=0
Mscore=0
nm=0
MaxM=0
while True:
    Type=input('Type of student (b/m):')
    if Type=='b':
        avg,nC=input().split()
        aa=float(avg)
        bb=int(nC)
        Score=getBSc(aa,bb)
        Bscore+=Score
        nb+=1
        if MaxB<Score:
            MaxB=Score
    elif Type=='m':
        avg,nP=input().split()
        avg=float(avg)
        nP=int(nP)
        Score=getMSc(avg,nP)
        Mscore+=Score
        nm+=1
        if MaxM<Score:
            MaxM=Score
    else:
        break

print('\nBachelor Students:')
print('Max Score is:',MaxB)
print('Average of scores is:',Bscore/nb)
print('\nMaster Students:')
print('Max Score is:',MaxM)
print('Average of scores is:',Mscore/nm)
```

## Example 5: Computing power

```python
def getPower(n,m):
    if n<m:
        n,m=m,n
    i=1
    P=1
    for i in range(m):
        P*=n
    return P
n=int(input('>'))
for i in range(n):
    a,b=input().split()
    a=int(a)
    b=int(b)
    r=getPower(a,b)
    print(r)
```
